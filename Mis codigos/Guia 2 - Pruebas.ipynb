```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
import random as rm
import math
```

```python
x = np.linspace (-1,5,1000)
r = -4
a = 4
f = lambda x: r* x + a * (x**2) - x**3
plt.plot (x,f(x),linestyle='dashed')
plt.xlabel ("x")
plt.ylabel ("f(x) \n Campo Vector")
plt.grid ()
plt.show ()
```

```python
r1 = np.linspace (-5,-2,100)
r2 = np.linspace (2,5,100)
x_fijo_rama1 = lambda r: (-r + ((r)**2 - 4)**(1/2))/2
x_fijo_rama2 = lambda r: (-r - ((r)**2 - 4)**(1/2))/2

## ------------------ ** ------------------ ##
plt.plot (r1,x_fijo_rama1 (r1))
plt.plot (r1,x_fijo_rama2(r1),linestyle='dashed')
plt.plot (r2,x_fijo_rama1 (r2),linestyle='dashed')
plt.plot (r2,x_fijo_rama2(r2))

## ------------------ ** ------------------ ##
plt.xlabel ("r")
plt.ylabel ("Valor del punto fijo \n en función del parámetro r")
plt.grid ()
plt.show ()
```

```python
r = np.linspace (-8,500,100+0)
a = 8
x_rama1 = lambda r: (-2*a + (4*a**2+12*r)**(1/2))/-6
x_rama2 = lambda r: (-2*a - (4*a**2+12*r)**(1/2))/-6
plt.plot (r,x_rama1 (r))
plt.plot (r,x_rama2 (r))
plt.grid ()
plt.show ()
```
